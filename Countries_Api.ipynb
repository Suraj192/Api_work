```python
import pandas as pd
import numpy as np
import requests
```

```python
response = requests.get("https://restcountries.eu/rest/v2/all")
response.status_code
```

```python
response.text
```

```python
data = response.json()
data = pd.DataFrame(data)
```

```python
pd.set_option('display.max_column', None)
data.head(3)
```

# Extracting Required Column

```python
currency_name = []
currency_symbol = []
for i in data["currencies"]:
    currency_name.append(i[0]["name"])
    currency_symbol.append(i[0]["symbol"])
    
currency_name = pd.DataFrame(currency_name)
currency_symbol = pd.DataFrame(currency_symbol)
currency_name = currency_name.rename(columns={0:'currency_name'})
currency_symbol = currency_symbol.rename(columns={0:'currency_symbol'})
print(currency_name.head(3))
print(currency_symbol.head(3))
```

```python
language = []
for i in data["languages"]:
    language.append(i[0]["name"])
language= pd.DataFrame(language)
language = language.rename(columns = {0:'main_language'})
language.head(3)
```

```python
data.columns
```

# Selecting required column from original data

```python
df = data.drop(['topLevelDomain', 'alpha2Code', 'altSpellings','demonym','gini','borders',
       'nativeName', 'numericCode', 'currencies', 'languages', 'translations',
       'flag', 'regionalBlocs', 'cioc'], axis = 1)
```

```python
df.head(3)
```

# Joining Currency and Language

```python
df1 = pd.concat([df, language, currency_name,currency_symbol], axis=1)
df1.head(3)
```

<div style='background-color:black; color:white; padding:25px'>
    Followed to the above description, we have empty list in the callingCodes and latlng which is not null value but also, it is not the value. Let's fix this one first. 
</div>

```python
latlngs = []
for i in df["latlng"]:
    if i == list():
        latlngs.append([5.883333,-162.083333])
    else:
        latlngs.append(i)
```

```python
latlngs = pd.DataFrame(latlngs).rename(columns = {0:"Latitude", 1:"Longitude"})
```

```python
df = df.join(latlngs, on = df.index, how="left")
```

```python
df.drop("latlng", axis=1,inplace=True)
```

```python
df.head()
```

```python
df[33:35]
```

<div style='background-color:black; color:white; padding:25px'>
    We have fix the empty list of latitude and longitude with proper latitude and longitude extracted from wikipedia.
</div>

```python
df["Countrycode"]=df.callingCodes.apply(lambda x:x[0])
```

```python
df.drop("callingCodes", inplace=True, axis=1)
```

```python
df["Countrycode"] = df.Countrycode.replace("", np.nan)
df["Countrycode"] = df.Countrycode.replace("1 340", "1340")
```
